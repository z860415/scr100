## 作業目標

    1. 取出知乎問題發問時間
    2. 取出第一筆與最後一筆回答的時間


### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間

import json
import requests 
import time

myheaders={"user-agent":"001"}
response=requests.get("https://www.zhihu.com/api/v4/questions/55493026/answers",headers=myheaders)
r=response.text
r

data=json.loads(r)
data

#取出問題發問時間

for d in data["data"]:
    createdtime=d["created_time"]
    answer=time.ctime(createdtime)
    print(answer)
#目前知乎上面的Createdtime為秒數，會需要將秒數轉換成正確數字需要使用到的為time.ctime)
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

firstanswer=time.ctime(data["data"][0]["updated_time"])
print("firstanswer is",firstanswer)
lastanswer=time.ctime(data["data"][-1]["updated_time"])  #-1為最後一筆基料，使用「-1」來去進行
print("lastanswer is", lastanswer)
# lastanswer=(f"lastanswer:{}")
```
