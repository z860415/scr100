1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json

url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
dt = requests.get(url)
data=json.loads(dt.text)#回傳jSON格式
i=0
for d in data:
    print(d["title"])
    i=i+1
print (i,"筆資料")
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

print("回傳",len(data[1]),"筆資料")
print(type(data[1]))
print("全部資料",data[1])
print("AllKey---->>>",data[1].keys())
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

count=0
while (True):
    try:
        each_data=data[count]
        print("標題: ",each_data["title"]," 貼文時間: ",each_data["createdAt"]," 留言人數: ",each_data["commentCount"]," 按讚人數: ",each_data["likeCount"])
        count=count+1
    except:
        break
```
